```python
import pandas as pd
import numpy as np
```

```python
# Concatenate all the dataframes of DSCOVR
dscovr_data = pd.read_csv("dscovrNEW/dsc_fc_summed_spectra_2016_v01.csv")

for path in range(2017, 2024):
    new_dscovr = pd.read_csv("dscovrNEW/dsc_fc_summed_spectra_"+ str(path)+"_v01.csv")
    dscovr_data = pd.concat([dscovr_data, new_dscovr], ignore_index=True)

print(dscovr_data.shape)

```

```python
# Concatenate all the dataframes of the validation data
valid_data = pd.read_csv("valid/omni2016.csv")

for path in range(2017, 2024):
    new_valid = pd.read_csv("valid/omni"+ str(path)+".csv")
    valid_data = pd.concat([valid_data, new_valid], ignore_index=True)

valid_data.rename(columns = {'YEAR': 'year', 'DAY': 'day', 'HOUR': 'hour'}, inplace = True)
```

```python
# Merging with the validation data
dscovr_data.merge(valid_data, on=['year', 'day', 'hour'], how='left', indicator=True)
```
